```python
data = load_data(r'C:\Users\VenEn\OneDrive\Desktop\python39\UCLA\ECE247\project')
X_train, y_train, person_train = data['X_train_valid'], data['y_train_valid'], data['person_train_valid']
X_test, y_test, person_test = data['X_test'], data['y_test'], data['person_test']
person_train = person_train[:, 0]
person_test = person_test[:, 0]
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# train data histograms
train_df = pd.DataFrame(data={'label': y_train, 'target': person_train})
train_df.hist()
plt.show()
```

```python
# test data histograms
test_df = pd.DataFrame(data={'label': y_test, 'target': person_test})
test_df.hist()
plt.show()
```

```python
def plot_ex(x, save_name=None):
    plt.figure(figsize=(20, 5))
    plt.imshow(x, cmap='gray')
    if save_name is not None:
        plt.savefig(fname=save_name)
    plt.show()

# %matplotlib qt
%matplotlib inline
for i in range(10):
    plot_ex(X_train[i])
```

```python
# stats
stats = pd.DataFrame()
stats["Channel"] = np.arange(X_train.shape[1])
stats["Mean"] = np.mean(X_train, axis=(-1, -3))
stats["Std"] = np.std(X_train, axis=(-1, -3))
stats["Max"] = np.max(X_train, axis=(-1, -3))
stats["Min"] = np.min(X_train, axis=(-1, -3))
print("X_train stats")
stats
```

```python
# classwise plots
s0_x_train = X_train[person_train == 0]
s0_y_train = y_train[person_train == 0]

# for each class, display 3 examples
classes = np.unique(y_train)
for c in classes:
    c_inds = np.arange(len(s0_x_train))[s0_y_train == c]
    samp = np.random.choice(c_inds, size=5, replace=False)
    print(f"Class {c}")
    for s in samp:
        plot_ex(s0_x_train[s])
```
